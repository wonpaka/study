```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
mtcars = pd.read_csv('./Documents/mtcars.csv')
```

```python
mtcars.head()
```

```python
mtcars.rename(columns= {'Unnamed: 0':'Name'}, inplace=True)
mtcars.head()
```

```python
mtcars = mtcars.set_index('Name')
mtcars.head()
```

```python
def min_max_scaler(df):
    df_min = df.min()
    df_max = df.max()

    return (df-df_min)/(df_max-df_min)
```

```python
mtcars['minmax'] = round(min_max_scaler(mtcars['qsec']),3)
mtcars.head()
```

```python
answer = mtcars[mtcars['minmax'] >= 0.5][mtcars[mtcars['minmax'] >= 0.5]['am'] == 0]
```
